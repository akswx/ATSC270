```python
'''
Example of lines plots for ATSC 270

1.Reads in text files of 2 m temperature data 
2.Each text file has GFK temp for each hour from 1 UTC - 23 UTC on 2/21/22 
for a totoal of 23 data points 
3.Text files must be in same directory as code (unless code is updated)

Written: Aaron Scott, UND Spring 2022
Last updated: 2/23/22
'''
import matplotlib.pyplot as plt

#read in data in text files 
#hrrr temp at 2 m in Fahrenheit 
hrrr_temp = [] #create empty list 
with open('hrrr_temp.txt','r') as f: #open text file as a file-object 
    for line in f: #loop through each line of file-object 
        hrrr_temp.append(float(line)) #convert to float and append to list 
#asos temp at 2 m in Fahrenheit 
asos_temp = []
with open('asos_temp.txt','r') as f:
    for line in f:
        asos_temp.append(float(line))
```

This link shows examples from the matplotlib documentation:<br>
<https://matplotlib.org/stable/plot_types/index>

```python
#method 1 
#uses the plot function from the pyplot submodule (of the matplotlib module)
#The plot function from pyplot creates the figure, and axes objects automatcially 

#generate list of UTC times that go with data 
UTC_times = [val for val in range(1,24,1)] #data starts at 1 UTC and ends at 23 UTC
plt.plot(UTC_times,asos_temp,color='blue') #UTC_times: x-coord, asos_temp: y-coord 
plt.title('ASOS Grand Forks, ND 2/21/22',size=15,fontweight='bold')
plt.ylabel('Temperature (deg F)',size=15,fontweight='bold')
plt.xlabel('Hour (UTC)',size=15,fontweight='bold')
#make ticks from 1-23 UTC at every other hour 
plt.xticks(range(1,24,2)) #range function returns sequence of numbers and passes to xticks function
plt.ylim(-15,-6) #set y limits 
plt.xlim(0,24) #set x limits 
plt.grid() #show grid 
```

What is the differnce in a function and a method in Python?
In simple terms, they will act very similiar and you will not really notice any difference.
However, pyplot.plot() is a function because pyplot is a module. <br>
Axes.plot() is a method because it belongs to an Axes object
We will not discuss objects in detail in this class, but you should at least know that we are creating objects. <br>
When we read in a file with the open() function, we are creating a file-object. Most things in Python are treated as objects.<br>
We will will explore objects a little more later in the class.

```python
#method 2 
#Manually create Figure object and Axes object (The OO-P style)

#generate list of UTC times that go with data 
UTC_times = [val for val in range(1,24,1)] #data starts at 1 UTC and ends at 23 UTC

fig = plt.figure(figsize=(8,4)) #create figure object and store in variable named fig;w/specified figure size
ax = fig.add_axes([0,0,1,1]) #create axes object attached to figure object;add_axes is method of figure object
ax.plot(UTC_times,asos_temp,color='blue') #UTC_times: x-coord, asos_temp: y-coord 
ax.set_title('ASOS Grand Forks, ND 2/21/22',size=15,fontweight='bold')
ax.set_ylabel('Temperature (deg F)',size=15,fontweight='bold')
ax.set_xlabel('Hour (UTC)',size=15,fontweight='bold')
#make ticks from 1-23 UTC at every other hour 
ax.set_xticks(range(1,24,2)) #range function returns sequence of numbers and passes to xticks function
ax.set_ylim(-15,-6) #set y limits 
ax.set_xlim(0,24) #set x limits 
ax.grid() #show grid 
```

Sometimes you may see code like below. <br>
This is creating the figure and axes objects in one line using the subplots() function from pyplot. <br>
The subplots() functions returns a tuple with the figure and axes objects. In this example, <br>
Python is unpacking the tuple and setting the figure object equal to fig and the axes objet equal to ax <br>
This is a good way and can be used in this class!

```python
fig, ax = plt.subplots()
```
