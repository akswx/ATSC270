```python
'''
Example of scatter plots for ATSC 270

1.Reads in text files of 2 m temperature data 
2.Each text file has GFK temp for each hour from 1 UTC - 23 UTC on 2/21/22 
for a totoal of 23 data points 
3.Text files must be in same directory as code (unless code is updated)

Written: Aaron Scott, UND Spring 2022
Last updated: 2/23/22
'''
import matplotlib.pyplot as plt

#read in data in text files 
#hrrr temp at 2 m in Fahrenheit 
hrrr_temp = [] #create empty list 
with open('hrrr_temp.txt','r') as f: #open text file as a file-object 
    for line in f: #loop through each line of file-object 
        hrrr_temp.append(float(line)) #convert to float and append to list 
#asos temp at 2 m in Fahrenheit 
asos_temp = []
with open('asos_temp.txt','r') as f:
    for line in f:
        asos_temp.append(float(line))
```

## Instructions for Group Exercise #2!
Change code below to plot night temps in blue and day temps in red <br>
Remember: each list (hrrr_temp and asos_temp) is 23 elements long. <br>
The first element is 1 UTC and the last element is 23 UTC. <br>
The frequencey of the data is 1 hour, so hrrr_temp[0]: 1 UTC, hrrr_temp[1]: 2 UTC, etc <br>
Think about how you would index the list for the night time and day time <br>
Hint: use index slicing! 

```python
#method 1 
#uses the plot function from the pyplot submodule (of the matplotlib module)
#The plot function from pyplot creates the figure, and axes objects automatcially 

#create scatter plot 
plt.scatter(hrrr_temp,asos_temp,color='blue',zorder=2) #change to plot only night temps (7pm-5am; 1 UTC - 11 UTC)
#add here to create plot of day temps (6am-5pm; 12 UTC - 23 UTC)
plt.ylim(-15,-6)
plt.xlim(-15,-6)
plt.xlabel('HRRR 2 m Temp (deg F)',size=15,fontweight='bold')
plt.ylabel('ASOS 2 m Temp (deg F)',size=15,fontweight='bold')
plt.title('Grand Forks, ND Feb 21,2022',size=15,fontweight='bold')
plt.grid()
plt.show()
```

```python
# Solution to above 
#method 1 
#uses the plot function from the pyplot submodule (of the matplotlib module)
#The plot function from pyplot creates the figure, and axes objects automatcially 

#create scatter plot - Note: [0:11] gives index 0-10 (1 UTC - 11 UTC) and [11::] gives index 11-22 [11 UTC - 23 UTC) Could also be [11:23]
plt.scatter(hrrr_temp[0:11],asos_temp[0:11],color='blue',zorder=2,label='night') #change to plot only night temps (7pm-5am; 1 UTC - 11 UTC)
plt.scatter(hrrr_temp[11::],asos_temp[11::],color='red',zorder=2,label='day')
#add here to create plot of day temps (6am-5pm; 12 UTC - 23 UTC)
plt.ylim(-15,-6)
plt.xlim(-15,-6)
plt.xlabel('HRRR 2 m Temp (deg F)',size=15,fontweight='bold')
plt.ylabel('ASOS 2 m Temp (deg F)',size=15,fontweight='bold')
plt.title('Grand Forks, ND Feb 21,2022',size=15,fontweight='bold')
plt.grid()
plt.show()
```

```python
# Solution to above 
#method 2
# Create fig object and axes object manually 

#create scatter plot - Note: [0:11] gives index 0-10 (1 UTC - 11 UTC) and [11::] gives index 11-22 [11 UTC - 23 UTC) Could also be [11:23]
fig = plt.figure() #creates figure object 
ax = fig.add_axes([.1,.1,.9,.9]) #creates axes object for our plot - axes bottom corner starts at 10% across and up page (figure object) - ends at 90% across and up
ax.scatter(hrrr_temp[0:11],asos_temp[0:11],color='blue',zorder=2,label='night') #change to plot only night temps (7pm-5am; 1 UTC - 11 UTC)
ax.scatter(hrrr_temp[11::],asos_temp[11::],color='red',zorder=2,label='day')

ax.set_ylim(-15,-6) #note on when using axes methods - there are some name differnces between these methods and the pyplot functions (see lecture)
ax.set_xlim(-15,-6)
ax.set_xlabel('HRRR 2 m Temp (deg F)',size=15,fontweight='bold')
ax.set_ylabel('ASOS 2 m Temp (deg F)',size=15,fontweight='bold')
ax.set_title('Grand Forks, ND Feb 21,2022',size=15,fontweight='bold')
ax.grid()
plt.show()
```
